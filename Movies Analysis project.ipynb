```python
my_data = my_data.fillna(0)
my_data['budget'] = my_data.budget.astype(int)
my_data['gross'] = my_data.gross.astype(int)
```

```python
my_data.head()
```

```python
my_data['CorrectYear'] = my_data['released'].astype(str).str[7:14]
```

```python
my_data.head()
```

```python
my_data.drop('CorrectYear',  inplace=True, axis=1)
```

```python
my_data.head()
```

```python
my_data['released'][1].split()
```

```python
my_data.drop('released', axis=1)
```

```python
my_data = my_data.sort_values(by='gross', inplace= False, ascending= False)
```

```python
plt.scatter(x=my_data['budget'], y=my_data['gross'])
plt.title('Budget vs Gross')
plt.xlabel('Budget')
plt.ylabel('gross')
plt.show
```

```python
sns.regplot(x= 'budget', y='gross', data= my_data, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})
```

```python
my_data.corr(method='pearson')
```

```python
corr_matrix = my_data.corr(method='pearson')
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation matrix')
plt.xlabel('Movies features')
plt.ylabel('Movies features')
plt.show()
```
